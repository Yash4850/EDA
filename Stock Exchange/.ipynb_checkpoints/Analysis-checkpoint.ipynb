```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pd.read_csv('raw data_sample.csv')
df.tail(10) 
```

# Data Preprocessing

```python
#converting datetime stamp to date and time
df['date_time'] = pd.to_datetime(df['datetime'])
df['date'] = df['date_time'].dt.date
df['time'] = df['date_time'].dt.time
df['day'] = df['date_time'].dt.day_name()
df.drop('datetime',axis =1, inplace = True) #dropping datetime
df.drop('date_time',axis =1, inplace = True) #dropping date_time
```

```python
#dropping application, node_name, category from their respective columns and keeping only the numbers
df['application_new'] = df['application'].str.split('_').str[1]
df['node_name_new'] = df['node_name'].str.split('_').str[2]
df['category_new'] = df['category'].str.split('_').str[1]
df.drop(['node_name','category','application'],axis = 1, inplace = True)
```

```python
df.tail() # final output
```

```python
# there are null values in category and application(espicially in application there are more null values)
df.info() 
```

```python
#converting values to numerical data
df['application_new'] = pd.to_numeric(df['application_new'])
df['node_name_new'] = pd.to_numeric(df['node_name_new'])
df['category_new'] = pd.to_numeric(df['category_new'])
```

```python
corr = df.corr() #for finding correlation
corr
```

```python
#As you can see the number of applications,node name and category are very high at lower values
sns.pairplot(df)
```

```python
# if you see here almost all the nan values in application are when node_name=6 and category=7
df[df.isnull().any(axis=1)] 
```

```python
 # we can confirm here with the number of rows 3478
df.loc[(df['node_name_new'] == 6) & (df['category_new'] == 7) & (df['application_new'].isna())]
```

```python
# we can see application=10 when nodename=6 and category=7
df.loc[(df['node_name_new'] == 6) & (df['category_new'] == 7) & (df['application_new'].notna())] 
```

```python
#So,I am replacing null values in application to 10
df['application_new']= df['application_new'].fillna(10)
```

```python
# if you see here almost all the nan values in category are when node_name=56 and application_new = 102 (or) node_name=0 and application_new = 10
df[df.isnull().any(axis=1)] 
```

```python
# if you see here when node_name=56 and application_new = 102 we have all the 87 rows and the day is always SATURDAY
df.loc[(df['application_new'] == 102) & (df['node_name_new'] == 56) & (df['category_new'].isna())]
```

Key Insights: 1)There is no 'Category' when 'application' is 102 and 'node_name' is 56 and it is on Saturday
2)Application number is missing when node_name is 6 and category is 7
